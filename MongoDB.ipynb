```python
# convert from JSON TO Python
#  if you have a JSON string , you can parse it by using the Jjson.loads() method
#  the result will be python dictionary
import json
# some JSON:
x = '{"name":"prashant","age":"24","city":"bhilai"}'
# parse x:
y = json.loads(x)
# the result is apyhton dictionary:
print(y)
print(y["age"])
```

```python
# convert from Python to JSON
# if you have a python objects , you can convert it in to a JSON string  by using the json.dumps() methods

#  a python object (dict)
x = {"name":"prashant", "age":"24","city":"bhilai"}
#   convert into json:
y = json.dumps(x)
# json is a string(result)
print(y)
```
